# Plotting with Plotnine in Python

```python
# Plotting Packages in Python

# - Matplot lib is most common (but clunky)

# - seaborn is less common, but much better

# - plotnine is much newer, and has EXACTLY the same syntax as ggplot2
```

```python
import numpy as np
import pandas as pd

from plotnine import *   # the * indicates "import everything"
```

# Simulate Some Data

```python
n = 1000

beta0 = 1.0
beta1 = 1.2
beta2 = 0.3

eps = np.random.normal(0, 2, n)      # simulate n number of observations from normal distribution which has mean 0 and std 2
x1 = np.random.normal(0, 2, n)
x2 = np.random.normal(0, 2, n)
x3 = np.random.choice(["g1", "g2"], n)   # simulate draws from a 2-group distribution

y = beta0 + beta1*x1 + beta2*x2 + eps
```

```python
df = pd.DataFrame({"x1": x1, "x2": x2, "x3": x3, "y": y})
```

```python
df.head()
```

# Scatter Plot

```python
(ggplot(df, aes("x1", "y"))
    + geom_point())
```

# Adding Color

```python
(ggplot(df, aes("x1", "y")) + geom_point(aes(color = "x2")))     # i.e. allow color to be determined by the value of x2
```

# Adding Layers

```python
(ggplot(df, aes("x1", "y")) + geom_point(aes(color = "x2")) + geom_smooth())    # color determined by x2, and adding a "smoother" i.e. a line
```

# Density Plot

```python
(ggplot(df, aes("x1")) + geom_density(color = "purple", fill = "yellow"))
```

```python
(ggplot(df, aes("x1", color = "x3", fill = "x3")) + geom_density())
```
